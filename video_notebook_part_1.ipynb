```python
import pandas as pd
import pyspark
from pyspark import SparkContext
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.functions import col
from IPython.display import clear_output

sc = SparkContext
spark = SparkSession.builder.appName("movie_guesser").getOrCreate()

spark.read.option("escape", "\"")
spark.conf.set("spark.sql.execution.arrow.enabled", "true")

csv_file_pandas = pd.read_csv("Movies.csv")
pyspark_csv = spark.createDataFrame(csv_file_pandas)
```

```python
points = 0
df = pyspark_csv.select(col("id"), col("title"), col("release_date"), col("vote_average"), col("vote_count"))
df.toPandas()
```

```python
def show_current_movies():
    print("first movie: ", movie_one["title"].values[0])
    print("second movie: ", movie_two["title"].values[0])



def get_two_random_movies():
    movie_one = df.toPandas().sample()
    movie_two = df.toPandas().sample()
    return movie_one, movie_two

movie_one, movie_two = get_two_random_movies()
show_current_movies()
```

```python
def show_correct_answer():
    print(f"first movie (", movie_one["title"].values[0], ") rating = ", movie_one["vote_average"].values[0])
    print(f"second movie (", movie_two["title"].values[0], ") rating = ", movie_two["vote_average"].values[0])

def check_guess(guess, answer):
    global points
    if(guess == answer):
        print("correct")
        points += 1
    else:
        print("incorrect")
        points = 0

def get_guesses(guess):
    global points
    global movie_one
    global movie_two
    clear_output(wait=True)
    vote_average_movie_one = float(movie_one["vote_average"])
    vote_average_movie_two = float(movie_two["vote_average"])
    
    if(vote_average_movie_one > vote_average_movie_two):
        check_guess(guess, 0)
    elif(vote_average_movie_one < vote_average_movie_two):
        check_guess(guess, 1)
    else:
        print("correct")
        points += 1
    
    show_correct_answer()
    print(f"your score is {points}")

    movie_one, movie_two = get_two_random_movies()
    print("")

```

```python
while(1==1):
    show_current_movies()
    guess = input()
    get_guesses(int(guess))
```
